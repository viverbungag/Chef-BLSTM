```python
from bs4 import BeautifulSoup
import requests
import pickle
import pandas as pd
import csv  
```

```python
df = pd.read_csv('../data/full_dataset.csv')
```

```python
csvRatingsPath = '../data/web_scraped/cookbook_ratings.csv'
```

```python
rating_limit = 4.0
```

```python
websites = set()
countDict = {}
for link in df["link"]:
  websites.add(link.split("/")[0])
  # if ("www.allrecipes.com" in link):
  #   counter += 1

for website in websites:
  countDict[website] = 0

for link in df["link"]:
  countDict[link.split("/")[0]] += 1
```

```python
cookbookLinks = []
for link in df["link"]:
  if ("www.cookbooks.com" in link):
    cookbookLinks.append(link)
```

```python
cookbookLinks
```

```python
try:
  df = pd.read_csv(csvRatingsPath)
  cookbookLinksLinksWithRatings = []
  cookbookLinksLinksWithRatings.extend(df.values.tolist())
except:
  cookbookLinksLinksWithRatings = []
```

```python
header = ['link', 'rating']
def getcookbookRatings(originalLink, idx):
  link = f'https://{originalLink[4:]}'
  try:
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'lxml')
    div = soup.find("div", {"id": "star-rating"})
    strong = div.find("strong")
    rating = strong.text
    cookbookLinksLinksWithRatings.append([originalLink, round(float(rating), 2)])
  except:
    cookbookLinksLinksWithRatings.append([originalLink, 0.0])

  if (idx % 100 == 0):
    print(idx)
    with open(csvRatingsPath, 'w') as f:
      write = csv.writer(f)
      write.writerow(header)
      write.writerows(cookbookLinksLinksWithRatings)

for idx in range(len(cookbookLinksLinksWithRatings), len(cookbookLinks)):
  getcookbookRatings(cookbookLinks[idx], idx)
```
