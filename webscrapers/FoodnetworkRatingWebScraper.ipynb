```python
from bs4 import BeautifulSoup
import requests
import pickle
import pandas as pd
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
import csv
```

```python
df = pd.read_csv('../data/full_dataset.csv')
```

```python
csvRatingsPath = '../data/web_scraped/foodnetwork_ratings.csv'
```

```python
links = []
for link in df["link"]:
  if ("www.foodnetwork.com" in link):
    links.append(link)
```

```python
links
```

```python
try:
  df = pd.read_csv(csvRatingsPath)
  linksWithRatings = []
  linksWithRatings.extend(df.values.tolist())
except:
  linksWithRatings = []
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless=new")
browser = Chrome(options=chrome_options)
header = ['link', 'rating', 'numberOfReviews']
def getRatings(originalLink, idx):
  link = f'https://{originalLink}'
  try:
    browser.get(link)
    soup = BeautifulSoup(browser.page_source, 'lxml')
    numberOfReviews = int(soup.find("h2", {"class": "reviews-ct"}).text.split(' ')[0])
    if (numberOfReviews > 0):
      div = soup.find("span", {"class": "sr-only"})
      rating = div.text.split(' ')[1]
      linksWithRatings.append([originalLink, round(float(rating), 2), numberOfReviews])
    else:
      linksWithRatings.append([originalLink, 0.0, numberOfReviews])
  except:
    linksWithRatings.append([originalLink, 0.0, 0])

  if (idx % 100 == 0):
    print(idx)
    with open(csvRatingsPath, 'w') as f:
      write = csv.writer(f)
      write.writerow(header)
      write.writerows(linksWithRatings)

for idx in range(len(linksWithRatings), len(links)):
  getRatings(links[idx], idx)

browser.quit()
```
