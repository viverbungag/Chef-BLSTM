```python
from bs4 import BeautifulSoup
import requests
import pickle
import pandas as pd
import csv  
```

```python
df = pd.read_csv('../data/full_dataset.csv')
```

```python
csvRatingsPath = '../data/web_scraped/tastykitchen_ratings.csv'
```

```python
rating_limit = 4.0
```

```python
websites = set()
countDict = {}
for link in df["link"]:
  websites.add(link.split("/")[0])
  # if ("www.allrecipes.com" in link):
  #   counter += 1

for website in websites:
  countDict[website] = 0

for link in df["link"]:
  countDict[link.split("/")[0]] += 1
```

```python
tastykitchenLinks = []
for link in df["link"]:
  if ("tastykitchen.com" in link):
    tastykitchenLinks.append(link)
```

```python
tastykitchenLinks
```

```python
try:
  df = pd.read_csv(csvRatingsPath)
  tastyKitchenLinksWithRatings = []
  tastyKitchenLinksWithRatings.extend(df.values.tolist())
except:
  tastyKitchenLinksWithRatings = []
```

```python
counter = 0
header = ['link', 'rating']
def getTastyKitchenRatings(originalLink):

  try:
    global counter
    link = f'https://{originalLink}'
    counter += 1
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'lxml')
    div = soup.find("span", {"class": "average"})
    if div:
      rating = div.find("a").text.split(" ")[0]
      tastyKitchenLinksWithRatings.append([originalLink, round(float(rating), 2)])
    else:
      tastyKitchenLinksWithRatings.append([originalLink, 0.0])
  except:
    tastyKitchenLinksWithRatings.append([originalLink, 0.0])
    
  if (counter % 100 == 0):
    print(counter)
    with open(csvRatingsPath, 'w') as f:
      write = csv.writer(f)
      write.writerow(header)
      write.writerows(tastyKitchenLinksWithRatings)

for idx in range(len(tastyKitchenLinksWithRatings), len(tastykitchenLinks)):
  getTastyKitchenRatings(tastykitchenLinks[idx])
```
