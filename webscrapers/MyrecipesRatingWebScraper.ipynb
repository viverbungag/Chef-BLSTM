```python
from bs4 import BeautifulSoup
import requests
import pickle
import pandas as pd
import csv  
```

```python
df = pd.read_csv('../data/full_dataset.csv')
```

```python
csvRatingsPath = '../data/web_scraped/myrecipes_ratings.csv'
```

```python
websites = set()
countDict = {}
for link in df["link"]:
  websites.add(link.split("/")[0])
  # if ("www.allrecipes.com" in link):
  #   counter += 1

for website in websites:
  countDict[website] = 0

for link in df["link"]:
  countDict[link.split("/")[0]] += 1
```

```python
myrecipesLinks = []
for link in df["link"]:
  if ("www.myrecipes.com" in link):
    myrecipesLinks.append(link)
```

```python
try:
  df = pd.read_csv(csvRatingsPath)
  myrecipesLinksWithRatings = []
  myrecipesLinksWithRatings.extend(df.values.tolist())
except:
  myrecipesLinksWithRatings = []
```

```python
header = ['link', 'rating']
def getMyrecipesRatings(originalLink, idx):
  link = f'https://{originalLink}'
  try:
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'lxml')
    lis = soup.findAll("li", {"class": "rating"})
    total_rating_count = 0
    total_rating = 0
    for li in lis:
      rating_stars = int(li.find("span", {"class": "rating-stars"}).text[0])
      rating_count = int(li.find("span", {"class": "rating-count"}).text)
      current_rating = rating_stars * rating_count
      total_rating += current_rating
      total_rating_count += rating_count

    rating = round(total_rating / total_rating_count, 2)

    myrecipesLinksWithRatings.append([originalLink, rating])
  except:
    myrecipesLinksWithRatings.append([originalLink, 0.0])

  if (idx % 100 == 0):
    print(idx)
    with open(csvRatingsPath, 'w') as f:
      write = csv.writer(f)
      write.writerow(header)
      write.writerows(myrecipesLinksWithRatings)


for idx in range(len(myrecipesLinksWithRatings), len(myrecipesLinks)):
  getMyrecipesRatings(myrecipesLinks[idx], idx)
```
