```python
from bs4 import BeautifulSoup
import requests
import pickle
import pandas as pd
```

```python
df = pd.read_csv('data/full_dataset.csv')
```

```python
websites = set()
countDict = {}
for link in df["link"]:
  websites.add(link.split("/")[0])
  # if ("www.allrecipes.com" in link):
  #   counter += 1

for website in websites:
  countDict[website] = 0

for link in df["link"]:
  countDict[link.split("/")[0]] += 1
```
